```python
import subprocess
from pathlib import Path

ROOT_DIR = Path("../data").resolve() 
DATA_DIR = ROOT_DIR / "LC/IGBP"
RAW_DIR = DATA_DIR / "rawdata"

FILLVALUE = "0"
raw_path = RAW_DIR / "global_igbp_15s_2020.nc.tif"
tmp1_path = DATA_DIR / "global_igbp_15s_2020.nc"
subprocess.run([
            "gdal_translate",
            "-q",
            "-of", "netCDF",
            "-co", "FORMAT=NC4",
            "-unscale",
            "-ot", "Float32",
            "-a_nodata", FILLVALUE,
            "-a_srs", "EPSG:4326",
            raw_path,
            str(tmp1_path),],
        check=True
    )
```
