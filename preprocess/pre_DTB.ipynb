### 1.Translate the 1km m DTB to 500m cm DTB 

```python
import subprocess
from pathlib import Path

ROOT_DIR = Path("../data").resolve()
DATA_DIR = ROOT_DIR / "DTB/Pelletier"
RAW_DIR  = DATA_DIR / "rawdata"

in_tif   = RAW_DIR / "average_soil_and_sedimentary-deposit_thickness.tif"
RES = "500m"
RES_DEG   = "0.0041666667"     

tmp_warp = DATA_DIR / f"DTB_Pelletier_{RES}_m_tmp.tif"     
tmp_cm   = DATA_DIR / f"DTB_Pelletier_{RES}_cm_tmp.tif"   
out_nc   = DATA_DIR / f"DTB_Pelletier_{RES}_cm.nc"

RESAMPLE  = "bilinear"              
SRC_NODATA = "-1"              
DST_NODATA = "-9999"

NUM_THREADS = "12"
ZLEVEL = "3"

subprocess.run([
    "gdalwarp",
    "-overwrite",
    "-t_srs", "EPSG:4326",
    "-te", "-180", "-90", "180", "90",
    "-tr", RES_DEG, RES_DEG,
    "-tap",
    "-r", RESAMPLE,
    "-multi",
    "-wo", f"NUM_THREADS={NUM_THREADS}",
    "-ot", "Float32",
    "-of", "GTiff",
    "-co", "TILED=YES",
    "-co", "COMPRESS=DEFLATE",
    "-co", f"ZLEVEL={ZLEVEL}",
    "-srcnodata", SRC_NODATA,
    "-dstnodata", DST_NODATA,
    str(in_tif), str(tmp_warp),
], check=True)

subprocess.run([
    "gdal_calc.py",
    "-A", str(tmp_warp),
    "--outfile", str(tmp_cm),
    "--calc", "A*100",
    "--type", "Float32",
    "--NoDataValue", DST_NODATA,
    "--co", "TILED=YES",
    "--co", "COMPRESS=DEFLATE",
    "--co", f"ZLEVEL={ZLEVEL}",
    "--quiet",
], check=True)

subprocess.run([
    "gdal_translate",
    "-q",
    "-of", "netCDF",
    "-co", "FORMAT=NC4",
    "-a_srs", "EPSG:4326",
    "-a_nodata", DST_NODATA,
    str(tmp_cm),
    str(out_nc),
], check=True)

print("[OK] written:", out_nc)
```
