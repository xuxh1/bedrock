```python
import numpy as np
import xarray as xr
from pathlib import Path

NOTEBOOK_DIR = Path().resolve()
ROOT_DIR = (NOTEBOOK_DIR / "../data").resolve()
DATA_DIR = ROOT_DIR / "Stocker"
RUN_DIR = ROOT_DIR / "run"

ds = xr.open_dataset(DATA_DIR / "cwdx80.nc")

# 构造全球 0.05° lat（中心点 -89.975..89.975）
lat_global = np.round(np.arange(-89.975, 90.0, 0.05), 3)  # -> 3600
# 保险：把原 lat 也 round 一下避免浮点对不上
ds = ds.assign_coords(lat=np.round(ds["lat"].values, 3))

# 直接在 lat 维度扩展，空出来的填 0
ds_global = ds.reindex(lat=lat_global, fill_value=np.nan)

ds_global.to_netcdf(DATA_DIR / "cwdx80_global.nc")
```
