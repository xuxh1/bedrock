### 1.Translate the 500m SAWS to p05 SAWS

```python
import subprocess
from pathlib import Path

NOTEBOOK_DIR = Path().resolve()
ROOT_DIR = (NOTEBOOK_DIR / "../data").resolve()
DATA_DIR = ROOT_DIR / "SAWS/Kosugi"
RAW_DIR  = DATA_DIR / "rawdata"

RES = "p05"


for i in range(1, 7):
    in_nc   = RAW_DIR / f"saws{i}.nc"
    out_nc   = DATA_DIR / f"SAWS_Kosugi_{RES}_m3m-3_{i}.nc"

    subprocess.run([
        "cdo","-O","-L", "-f", "nc4","-z","zip_3","-b","F32",
        "-P", "32", 
        f"remapbil,{ROOT_DIR}/p05.txt",
        str(in_nc),
        str(out_nc),
    ], check=True)

    print("[OK] written:", out_nc)
```
