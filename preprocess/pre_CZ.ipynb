```python
import subprocess
from pathlib import Path

ROOT_DIR = Path("../data").resolve()
DATA_DIR = ROOT_DIR / "CZ/Koppen"
RAW_DIR  = DATA_DIR / "rawdata"

in_tif  = RAW_DIR / "Beck_KG_V1_present_0p0083.tif"
tmp_tif = DATA_DIR / "CZ_Koppen_500m_tmp.tif"
out_nc  = DATA_DIR / "CZ_Koppen_500m_tmp.nc"

RES_DEG = "0.0041666667"
RESAMPLE = "near"
DST_NODATA = "-9999"
SRC_NODATA = "0"

NUM_THREADS = "12"
ZLEVEL = "3"

subprocess.run([
    "gdalwarp",
    "-overwrite",
    "-t_srs", "EPSG:4326",
    "-te", "-180", "-90", "180", "90",
    "-tr", RES_DEG, RES_DEG,
    "-tap",
    "-r", RESAMPLE,
    "-multi",
    "-wo", f"NUM_THREADS={NUM_THREADS}",
    "-ot", "Byte",
    "-of", "GTiff",
    "-co", "TILED=YES",
    "-co", "COMPRESS=DEFLATE",
    "-co", f"ZLEVEL={ZLEVEL}",
    "-srcnodata", SRC_NODATA, 
    "-dstnodata", DST_NODATA,
    str(in_tif), str(tmp_tif),
], check=True)

subprocess.run([
    "gdal_translate",
    "-q",
    "-of", "netCDF",
    "-co", "FORMAT=NC4",
    "-a_srs", "EPSG:4326",
    "-a_nodata", DST_NODATA,
    str(tmp_tif),
    str(out_nc),
], check=True)

print("[OK] written:", out_nc)
```
