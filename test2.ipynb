```python
import geopandas as gpd
import matplotlib.pyplot as plt

# ----------------------------------------
# 1. 设置文件路径 (请修改为你自己的shp文件路径)
# ----------------------------------------
# 注意：Windows路径如果包含反斜杠 \，请在字符串前加 r，例如 r"D:\data\test.shp"
ROOT_DIR = "/tera04/zhwei/xionghui/bedrock/data/Shp/"
# shp_path = f"{ROOT_DIR}China/chinawithisland.shp" 
shp_path = f"{ROOT_DIR}continent/continent_AF.shp" 


try:
    # ----------------------------------------
    # 2. 读取数据
    # ----------------------------------------
    # encoding参数处理中文乱码，常见的是 'utf-8' 或 'gbk' (即cp936)
    # 如果报错提示编码问题，尝试修改 encoding='gbk'
    gdf = gpd.read_file(shp_path, encoding='utf-8')

    # ----------------------------------------
    # 3. 查看基本信息 (可选)
    # ----------------------------------------
    print(f"数据加载成功！")
    print(f"坐标系 (CRS): {gdf.crs}")  # 查看坐标系，例如 EPSG:4326
    print(f"数据规模: {gdf.shape[0]} 个要素 (行), {gdf.shape[1]} 个属性 (列)")
    print("-" * 30)
    print("前5行数据预览:")
    print(gdf.head()) # 让你看看除了形状外，里面存了什么表格数据

    # ----------------------------------------
    # 4. 绘图 (可视化)
    # ----------------------------------------
    # figsize 控制图片大小 (宽, 高)
    fig, ax = plt.subplots(figsize=(10, 10))
    
    # 绘制图层
    # facecolor: 填充颜色 (例如 'white', 'blue', 'none')
    # edgecolor: 边框颜色 (例如 'black', 'red')
    # linewidth: 边框粗细
    gdf.plot(ax=ax, facecolor='lightblue', edgecolor='black', linewidth=0.5)

    # 设置标题
    ax.set_title("SHP File Visualization", fontsize=15)
    
    # 移除坐标轴刻度 (如果你只想看形状，不关心经纬度数字)
    # ax.set_axis_off() 

    # 显示图片
    plt.show()

except Exception as e:
    print(f"发生错误: {e}")
```
