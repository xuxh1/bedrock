### 2. Preprocess the mask all Dbedrock > 0 

```python
import subprocess
from pathlib import Path
from concurrent.futures import ProcessPoolExecutor

ROOT_DIR = Path("/home/xuxh22/stu01/bedrock/data") 
in_dir = ROOT_DIR / "yearly_Dr_Dbedrock"
out_dir = ROOT_DIR / "mask/mask_Dbedrock"
(out_dir / "ind").mkdir(parents=True, exist_ok=True)

files = sorted(in_dir.glob("Dbedrock_*.nc"))

def one_file(f):
    year = f.stem.split("_")[-1]  # Dbedrock_2003.nc -> "2003"
    out = out_dir / "ind" / f"Dbedrock_gt0_{year}.nc"
    subprocess.run(
        ["cdo", "-O", "-L", "-f", "nc4", "-z", "zip_3",
         "-setmisstoc,0", "-gtc,0", str(f), str(out)],
        check=True
    )
    return str(out)

with ProcessPoolExecutor(max_workers=24) as ex:
    ind_files = list(ex.map(one_file, files))

subprocess.run(
    ["cdo", "-O", "-L", "-f", "nc4", "-z", "zip_3",
     "ensmin", *ind_files, str(out_dir / "mask_Dbedrock.nc")],
    check=True
)

print("done:", out_dir / "mask_Dbedrock.nc")
```
